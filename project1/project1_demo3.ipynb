```python
# define the y' funciotn
def oscillator(t,y,K,M):
        """
        The derivate function for an oscillator
        In this example, we set

        y[0] = x
        y[1] = v

        yderive[0] = x' = v
        yderive[1] = v' = a

        :param t: the time
        :param y: the initial condition y
        :param K: the spring constant
        :param M: the mass of the oscillator

        """
        omega0 = np.sqrt(K/M)
        y = np.array([y[1], -omega0**2 * y[0]])
 
        return y # <- change here. just a placeholder

# Prepare the input arguments
t_span = (0, 10)
y0     = np.array([1,0])
t_eval = np.linspace(0,10,1000)

K = 1
M = 1

# use the IVP solver

mySol = mysolver.solve_ivp(oscillator, t_span, y0, 
                    method="RK4",t_eval=t_eval, args=(K,M))

```

```python
# visualize the results

plt.figure
plt.plot(t_eval,mySol[0,:])
plt.xlabel('t(sec)')
plt.title('1-D SHO')
plt.show()
```

# Damped Oscillation

When there is a damping force ($F_{\rm damp}=-\lambda \dot{x}$), the equation of motion becomes,

$$
m\ddot{x} + \lambda \dot{x} + kx = 0
$$

### Analytical Solutions

The analytical solution is

$$
x(t) = A e^{-\gamma t} \left[\cos(\omega t + \phi) \right],
$$

where $\omega = \sqrt{\gamma^2 -\omega_0^2}$ or $\omega=\sqrt{\omega_0^2-\gamma^2}$.

```python
# Compute the analytical solution

A = 1
gamma = L/(2*M)
omega0 = np.sqrt(K/M)
omega = np.sqrt(omega0**2 - gamma**2)

X = A* np.exp(-gamma*t_eval)*np.cos(omega*t_eval)

plt.figure
plt.plot(t_eval,mySol[0])
plt.plot(t_eval,X)
plt.xlabel('t(sec)')
plt.title('1-D SHO')
plt.show()

```
